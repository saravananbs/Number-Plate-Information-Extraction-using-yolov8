```python
from ultralytics import YOLO

# Load the YOLOv8 model (choose 'yolov8n', 'yolov8s', etc., based on your needs)
model = YOLO("yolov8n.pt")  # Use 'n' for nano, 's' for small, etc.

# Train the model
model.train(
    data="/content/License-Plate-Recognition-4/data.yaml",  # Path to the dataset's YAML file
    epochs=50,                # Number of training epochs
    imgsz=640,                # Image size
    batch=16,                 # Batch size
    device=0                  # Set to '0' for GPU or 'cpu' for CPU
)
```

```python
metrics = model.val(data="/content/License-Plate-Recognition-4/data.yaml")
print(metrics)
```

```python
# Load the trained model (after training, a .pt file is saved in the 'runs/train' directory)
trained_model = YOLO("/content/yolo11n.pt")

# Run inference
results = trained_model.predict("/content/License-Plate-Recognition-4/test/images/CarLongPlate324_jpg.rf.0ca516f05861b6afd26480d4051af1d6.jpg", save=True)

# Print results
print(results)
```
